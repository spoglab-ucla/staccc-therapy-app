### Performing sanity checks on ALICE performance on the first 30 mins of 032LTP2 and 001LTP1


 Clipping out sections from wav file to listen to:

 Data composition: 20 clips of adult speech + 80 clips of child speech

 clip format: ALICEtrial_startonset_endoffset_CHI

 csv output: all these file names with their count outputs from ALICE

```python
utterance_counts = pd.read_csv('ALICE/ALICE_output_utterances_alicetrial2.txt', sep="\t", header=None)
```

```python
utterance_counts.columns=['file','phoneme', 'syllable', 'word']
utterance_counts
```

```python
def extract_timestamps(name):
    lastunderscore = name.rfind("_")
    time_end = name[lastunderscore+1:lastunderscore+11]
    secondlastunderscore = name[:lastunderscore].rfind("_")
    time_start = name[secondlastunderscore+1: lastunderscore]
    return time_start, time_end

def extract_start_time(name):
    start, end = extract_timestamps(name)
    return start

def extract_end_time(name):
    start, end = extract_timestamps(name)
    return end
```

```python
utterance_counts['start_time'] = utterance_counts['file'].apply(extract_start_time)
utterance_counts['end_time'] = utterance_counts['file'].apply(extract_end_time)
```

```python
utterance_counts
```

```python
diarization  = pd.read_csv('ALICE/diarization_output_alicetrial2.rttm', sep=" ", header=None)
```

```python
mask = diarization[7] == 'SPEECH'
non_speech_diarized = diarization[~mask][7]
```

```python
utterance_counts['speaker'] = non_speech_diarized.reset_index(drop=True)
```

```python
utterance_counts
```

```python
childclips = 0
adclips = 0
row = 0

wavpath = "/Users/arjun/Box/box-group-lena-studies/Soundscape/wavFiles/sample_expt_arjun/alicetrial2.wav"
rate, wavdata = wavfile.read(wavpath)

while (True):
    if (adclips == 20 and childclips == 80):
        break

    if (adclips >=20 and (utterance_counts['speaker'][row] == 'MAL' or utterance_counts['speaker'][row] =='FEM')):
        row = row+1
        continue

    if (childclips >=80 and (utterance_counts['speaker'][row] == 'KCHI' or utterance_counts['speaker'][row] =='CHI')):
        row = row+1
        continue

    start = float(utterance_counts['start_time'][row])/10000
    ending = float(utterance_counts['end_time'][row])/10000
    split_at_frame = math.floor(rate * start)
    stop_at_frame = math.ceil(rate * ending)
    left_data  = wavdata[split_at_frame : stop_at_frame]
    clipname = utterance_counts['speaker'][row]+"_" + str(start)+"_"+str(ending)+"p-"+str(utterance_counts['phoneme'][row])+"sy-"+str(utterance_counts['syllable'][row])
    newpath = os.path.join("/Users/arjun/Box/box-group-lena-studies/Soundscape/ucsf_ALICE_expts/alicetrial2results_001LTP1", clipname+".wav")
    wavfile.write(newpath, rate, left_data)
    
    if utterance_counts['speaker'][row] == 'KCHI' or utterance_counts['speaker'][row] =='CHI':
        childclips = childclips+1
    else:
        adclips = adclips+1
    
    row = row+1
```

```python
utterance_counts['start_time(in sec)'] = pd.to_numeric(utterance_counts['start_time']).apply(lambda x: x/10000)
utterance_counts['end_time(in sec)']  = pd.to_numeric(utterance_counts['end_time']).apply(lambda x: x/10000)
utterance_counts.to_excel("/Users/arjun/Box/box-group-lena-studies/Soundscape/ucsf_ALICE_expts/merged_ALICE_output.xlsx")
```
